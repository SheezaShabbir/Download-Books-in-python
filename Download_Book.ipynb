# Search by title

```python
s = LibgenSearch()
results = s.search_title("Deep Learning for Vision Systems")

# As the above output will be in JSON format so create a dataframe for the results
# use json_normalize to create a dataframe using JSON value

result_df = pd.json_normalize(results)
pd.set_option("display.max_rows",result_df.shape[0]+1)
result_df.head(2).T # to transpose the dataframe
```

## Fitered Searching by title

```python
tf = LibgenSearch()
title_filters = {"Year": "2020", "Extension": "pdf"}
results = tf.search_title_filtered("Deep Learning for Vision Systems", title_filters, exact_match=True)
results[0]
```

## Fitered Searching by author

```python
af = LibgenSearch()
author_filters = {"Language": "English", "Year": "2020"}
results = af.search_author_filtered("Mohamed Elgendy", author_filters, exact_match=True)
results[0]
```

## Non-Exact Filtered Searching- (exact_match = False)

```python
ne_af = LibgenSearch()
partial_filters = {"Year": "2020"}
results = ne_af.search_author_filtered("Mohamed Elgendy", partial_filters, exact_match=False)
results[0]
```

```python
results[0]
```

# Downloading Based on Book ID

```python
item_to_download = results[0]
download_links = ne_af.resolve_download_links(item_to_download)
links = pd.json_normalize(download_links)
links.T
```
